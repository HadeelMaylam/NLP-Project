```python
import pandas as pd
```

```python
import string
```

```python
df = pd.read_csv('series_data.csv')
df.info()
```

```python
df.head()
```

```python
# remove null overview and name
df = df[df['name'].notnull()]
df = df[df['overview'].notnull()]
df.info()
```

```python
df['text'] = df['overview']
```

```python
# To lowercase
df['text'] = df['text'].apply(lambda s: s.lower())
df.head()
```

```python
# Remove puncutation
df['text'] = df['text'].apply(lambda s: s.translate(str.maketrans('', '', string.punctuation)))
df.head()
```

```python
import re
```

```python
df['text'] = df['text'].apply(lambda s: re.sub(r'[^a-zA-Z\s]', '', s))
df.head()
```

```python
# Remove Repetitive
df['text'] = df['text'].apply(lambda s: re.sub(r'\b\w*(\w)\1{2,}\w*\b', '', s))
df.head()
```

```python
import nltk
from nltk.corpus import stopwords
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
```

```python
# Remove stopwords
df['text'] = df['text'].apply(lambda s: ' '.join(word for word in s.split() if word.lower() not in stop_words))
df.head()
```

```python
# Remove single characters words
df['text'] = df['text'].apply(lambda s: re.sub(r'\b\w\b', '', s).strip())

# Remove single word rows
df = df[df['text'].str.split().apply(len) > 1]
df.reset_index(drop=True, inplace=True)
df
```

```python
import spacy

# Load spaCy language model
nlp = spacy.load("en_core_web_sm")

# Lemmatization function
def lemmatize_text_spacy(text):
    doc = nlp(text)  # Process the text
    return ' '.join([token.lemma_ for token in doc])  # Extract lemmatized words

# Apply lemmatization
df['lemmatized_text'] = df['text'].apply(lemmatize_text_spacy)

```

```python
df['lemmatized_text'].head()
df['text'] = df['lemmatized_text']
```

## TF-IDF

```python
from sklearn.feature_extraction.text import TfidfVectorizer
```

```python
tfidf = TfidfVectorizer()
result = tfidf.fit_transform(df['text'])
```

```python
tfidf_df = pd.DataFrame(result.toarray(), columns=tfidf.get_feature_names_out())
tfidf_df.info()
```

```python
pd.concat([tfidf_df, df[['name', 'slug']].reset_index(drop=True)], axis=1).to_csv('tfidf_serieses.csv')
```
